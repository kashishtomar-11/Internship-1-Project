```python
df_cleaned['InvoiceDate'] = pd.to_datetime(df_cleaned['InvoiceDate'])
```

## Transform the Data

```python
numerical_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
```

```python
df_cleaned[numerical_cols] = scaler.fit_transform(df_cleaned[numerical_cols])
```

```python
for col in categorical_columns:
    df_cleaned[col] = df_cleaned[col].astype(str)
```

```python
df_cleaned['InvoiceYear'] = df_cleaned['InvoiceDate'].dt.year
df_cleaned['InvoiceMonth'] = df_cleaned['InvoiceDate'].dt.month
df_cleaned['InvoiceDay'] = df_cleaned['InvoiceDate'].dt.day
df_cleaned['InvoiceHour'] = df_cleaned['InvoiceDate'].dt.hour
df_cleaned['InvoiceMinute'] = df_cleaned['InvoiceDate'].dt.minute
```

## Feature Engineering

```python
df_cleaned['TotalPrice'] = df_cleaned['Quantity'] * df_cleaned['UnitPrice']
```

```python
X = df_cleaned.drop(['TotalPrice'], axis=1)
y = df_cleaned['TotalPrice']
```

```python
selector = SelectKBest(score_func=f_classif, k=5)
X_new = selector.fit_transform(X, y)
```

```python
selected_features = X.columns[selector.get_support()]
```

```python
selected_features
```

## Standardize the Data

```python
features = ['Quantity', 'UnitPrice']
```

```python
scaler = StandardScaler()
df[features] = scaler.fit_transform(df[features])
```

```python
print(df[features].head())
```

## Determine the Optimal Number of Clusters using the Elbow Method

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df[features])
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

## The determination of the optimal number of clusters using the elbow method involves identifying the point where the within-cluster sum of squares (WCSS) starts to decrease more slowly.In your case, the decrease from ð‘˜=3 to k=4 is less steep compared to the decrease from k=2 to k=4 , suggesting that k=3 might be the optimal choice. 

## Run K-Means Clustering

```python
optimal_clusters = 3  
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[features])
```

```python
print(df['Cluster'].value_counts())
```

```python
print(df.head())
```

## Visualize the Clusters

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='UnitPrice', hue='Cluster', data=df, palette='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.title('Customer Segmentation')
plt.xlabel('Quantity')
plt.ylabel('UnitPrice')
plt.legend()
plt.show()
```
